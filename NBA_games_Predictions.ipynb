```python
from nba_api.stats.endpoints import leaguegamefinder
```

```python
# d'apres la documenttaion , nba id = 00 ( LIEN : https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/library/parameters.md#LeagueID)
gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable='01/31/2020' , league_id_nullable='00')
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games.shape
```

```python
games.columns
```

```python
games = games[['TEAM_NAME', 'GAME_ID','GAME_DATE', 'MATCHUP', 'WL' ,'PLUS_MINUS' ]]
```

```python
games
```

```python
# final dataframe ; one row for one game
#                   two columns : 1.result of game :target , 2.score stat comapring two teams: feature
```

```python
import pandas as pd
games.info()

# we need to convert the game_date type to date time format
```

```python
games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])
```

```python
games.GAME_DATE
```

```python
games = games.sort_values('GAME_DATE')
```

```python
games
```

```python
# calculate for each team the mean of plus minus score for the last 30 game
#closed= 'left' that mean without counting the current game 
games['avg_30_plus_minus'] = games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x:  x.rolling(30 , closed='left').mean())
```

```python
# here the first 30 row in the colone avg_30_plus_minus show NAN values because we needthe the avg of the last 30 game , for example for the row number 18 , we have not reach the game 30 to calculate the avg n fior that there is enght data 
games[games['TEAM_NAME']=='Toronto Raptors'].head(35)
```

```python
msk = games['MATCHUP'].str.contains('@')
```

```python
games_away = games[msk]
games_home = games[~msk]
```

```python
games_away.shape
```

```python
games_home.shape
```

```python
games_home
```

```python
games_away
```

```python
games_merged = pd.merge(games_home , games_away , on='GAME_ID' , suffixes=('_home' , '_away'))
```

```python
games_merged
```

```python
games_merged['avg_30_plus_minus_diff'] = games_merged['avg_30_plus_minus_home']-games_merged['avg_30_plus_minus_away']
```

```python
games_merged
```

```python
games_model = games_merged[['WL_home' , 'avg_30_plus_minus_diff']].dropna()
```

```python
games_model
```

```python
games_model['WL_home'] = games_model['WL_home'].map({'W':1 , 'L':0})
```

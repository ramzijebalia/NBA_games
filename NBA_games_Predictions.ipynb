```python
from nba_api.stats.endpoints import leaguegamefinder
```

```python
# d'apres la documenttaion , nba id = 00 ( LIEN : https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/library/parameters.md#LeagueID)
gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable='01/31/2020' , league_id_nullable='00')
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games.shape
```

```python
games.columns
```

```python
games = games[['TEAM_NAME', 'GAME_ID','GAME_DATE', 'MATCHUP', 'WL' ,'PLUS_MINUS' ]]
```

```python
games
```

```python
# final dataframe ; one row for one game
#                   two columns : 1.result of game :target , 2.score stat comapring two teams: feature
```

```python
import pandas as pd
games.info()

# we need to convert the game_date type to date time format
```

```python
games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])
```

```python
games.GAME_DATE
```

```python
games = games.sort_values('GAME_DATE')
```

```python
games
```

```python
# calculate for each team the mean of plus minus score for the last 30 game
#closed= 'left' that mean without counting the current game 
games['avg_30_plus_minus'] = games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x:  x.rolling(30 , closed='left').mean())
```

```python
# here the first 30 row in the colone avg_30_plus_minus show NAN values because we needthe the avg of the last 30 game , for example for the row number 18 , we have not reach the game 30 to calculate the avg n fior that there is enght data 
games[games['TEAM_NAME']=='Toronto Raptors'].head(35)
```

```python
msk = games['MATCHUP'].str.contains('@')
```

```python
games_away = games[msk]
games_home = games[~msk]
```

```python
games_away.shape
```

```python
games_home.shape
```

```python
games_home
```

```python
games_away
```

```python
games_merged = pd.merge(games_home , games_away , on='GAME_ID' , suffixes=('_home' , '_away'))
```

```python
games_merged
```

```python
games_merged['avg_30_plus_minus_diff'] = games_merged['avg_30_plus_minus_home']-games_merged['avg_30_plus_minus_away']
```

```python
games_merged
```

```python
games_model = games_merged[['WL_home' , 'avg_30_plus_minus_diff']].dropna()
```

```python
games_model
```

```python
games_model['WL_home'] = games_model['WL_home'].map({'W':1 , 'L':0})
```

```python
games_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train , df_test = train_test_split(games_model , stratify=games_model['WL_home'] , test_size=0.3 , random_state=7)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
target = 'WL_home'

x_train  =df_train.drop(columns=target)
y_train = df_train[target]

x_test  =df_train.drop(columns=target)
y_test = df_train[target]
```

```python
# we focsed on XDBoost to show the general steps of modeling
# XGBoost is a popular gradiant boosting libarie
```

```python
import xgboost as xgb
```

```python
clf = xgb.XGBClassifier(random_state=7)
clf.fit(x_train , y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test , y_pred)
```

triying to improve the model 

```python
#RandomizedSearchCV is a randomized search over a hyperparameter space,
# which means that it randomly selects hyperparameters from a specified
# distribution to evaluate the performance of the model. 
#It then uses cross-validation to evaluate the performance of the model
# on different subsets of the data, and returns the best hyperparameters
# that result in the best cross-validation performance.
```

```python
# hyperparameter : it is  a parameters with values was set before the learning process begins
# hyperparameter tuning - approch : ranndom search 
from sklearn.model_selection import RandomizedSearchCV
from sklearn.utils.fixes import loguniform
```

```python
hyp_params = {'learning_rate' : loguniform(0.0001 ,1), 
              'max_depth' :[1,2,3,4,5,6,7,8,9],
              'subsample' : [0.5,0.6,0.7,0.8,0.9,1.0],
              'n_estimators' : [25,50,75,100,125,150,200]}
```

```python
random_hyp = RandomizedSearchCV(estimator=clf,
                                param_distributions=hyp_params,
                                n_iter = 15,
                                cv=7,
                                scoring='accuracy',
                                random_state=7)
```

```python
random_hyp.fit(x_train , y_train)
```

```python
random_hyp.best_params_
```

```python
model_hyp = random_hyp.best_estimator_
```

```python
y_pred_hyp = model_hyp.predict(x_test)
```

```python
accuracy_score(y_test , y_pred_hyp)
```
